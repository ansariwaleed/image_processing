**SIFT - scale Invariant Feature Transform**
----
What is sift ? - a method to find feature in images.
--
why do we need SIFT ? -
1.   feature detection
2.   Robust

***How does Sift work ? ***
> Step 1: Scalespace extrema detection
1.   Finding blobs by approximation LoG and Dog
2.   Obtain gaussian pyramid and for each layer in a octave blur with different sigmas fro Dog. Then find extremes points



> Step 2 : keypoint Localization
1.   REmove points with low contrast since they are sensitive to noise
2.   Tries to find more accurate location of extrema using Taylor expansion of scale-space function and finding where D'(x) = 0

3. Rejects points with high edge response bu using the hessian (second dericative) because unstable to noise

> step 3 - Orientation Assignment
- find orietation(magnitude and angle) of the keypoint to be invariant to rotaition.
image.png

> Step 4 - Keypoint descriptor
- define the keypoint by creating a vector descibing the orentation histogram of the neghbours
image.png
> Step 5: Keypoitn matching
-find keypoint match using nearest neighbour and L2 distanece to find minimum error between hitogram vectors.

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def SIFT():
  root = os.getcwd()
  imgPath = os.path.join(root,'/content/drive/MyDrive/48c974edb20ba1823f6020416acd638a.jpg')
  imggray= cv.imread(imgPath,cv.IMREAD_GRAYSCALE)

  sift = cv.SIFT_create()
  keypoints = sift.detect(imggray,None)
  imgGray = cv.drawKeypoints(imggray,keypoints,imggray,flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

  plt.figure(figsize=(12, 8)) # Changed to plt.figure() to create a new figure
  plt.imshow(imgGray) # Added imgGray as argument to imshow to display the image
  plt.show()

if __name__ == '__main__':
  SIFT() # Added parentheses to call the function
```
